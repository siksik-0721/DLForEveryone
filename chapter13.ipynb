# 13장 모델 성능 검증하기

## 1. 데이터의 확인과 예측 실행

```python
# pandas 라이브러리를 불러옵니다.
import pandas as pd

# 광물 데이터를 불러옵니다.
df = pd.read_csv("./data/sonar3.csv", header=None)

df.head()  # 첫 다섯 줄을 봅니다.
```

```python
df[60].value_counts()
```

```python
X = df.iloc[:, 0:60]
y = df.iloc[:, 60]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 모델을 설정합니다.
model = Sequential()
model.add(Dense(24, input_dim=60, activation="relu"))
model.add(Dense(10, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

# 모델을 컴파일합니다.
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# 모델을 실행합니다.
history = model.fit(X, y, epochs=200, batch_size=10)
```

## 2. 과적합 이해하기

- 과적합(overfitting): 모델이 학습 데이터셋 안에서는 일정 수준 이상의 예측 정확도를 보이지만, 새로운 데이터에 적용하면 잘 맞지 않는 것
    - 층이 너무 많거나 변수가 복잡해서 발생 가능
    - 테스트셋과 학습셋이 중복될 때 발생 가능

## 3. 학습셋과 테스트셋

- 학습을 하는 데이터셋과 이를 테스트할 데이터셋을 완전히 구분
    - 학습과 동시에 테스트를 병행하며 진행하는 것이 과적합 방지의 한 방법
- 학습셋만 가지고 평가할 때, 층을 더하거나 에포크 값을 높이면 정확도가 계속 올라갈 수 있음
    - 하지만 테스트셋에서도 유사한 예측 성공률이 나타나지 않을 수 있음 -> 과적합
- 학습이 계속되면 학습셋에서의 에러는 계속해서 작아지지만, 테스트셋은 에러가 작아지다 다시 커지는 과적합이 발생
    - 학습을 진행해도 테스트 결과가 더 이상 좋아지지 않는 지점이 가장 적절

```python
from sklearn.model_selection import train_test_split

# 학습셋과 테스트셋을 구분합니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
# 모델을 설정합니다.
model = Sequential()
model.add(Dense(24, input_dim=60, activation="relu"))
model.add(Dense(10, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

# 모델을 컴파일합니다.
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# 모델을 실행합니다.
history = model.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
# 모델을 테스트셋에 적용해 정확도를 구합니다.
score = model.evaluate(X_test, y_test)
print("Test accuracy:", score[1])
```

- 딥러닝, 머신 러닝의 목표
    - 새로운 데이터에 대해 높은 정확도를 안정되게 보여 주는 모델을 만드는 것
    - 학습셋에서만 잘 작동하는 모델이 아님
- 모델 성능의 향상
    - 데이터를 보강
        - 충분한 데이터를 가져와 추가
            - 딥러닝의 경우 샘플 수가 많을수록 일반적으로 성능이 좋아짐
        - 가지고 있는 데이터를 적절히 보완
            - 너무 크거나 낮은 이상치가 영향을 주지 않도록 크기를 적절히 조절
            - 교차 검증 방법을 사용해 가지고 있는 데이터를 충분히 이용
    - 알고리즘을 최적화
        - 다른 구조로 모델을 바꾸며 최적화
            - 은닉층의 개수, 노드 수, 최적화 함수 종류 등
        - 딥러닝이 아닌 다른 알고리즘 사용

## 5. k겹 교차 검증

- 실제 프로젝트에서는 데이터 확보가 어려워 가지고 있는 데이터를 활용하는 것이 중요
    - 테스트셋 30% 설정 시, 해당 데이터들은 학습에 이용할 수 없다는 단점
- k겹 교차 검증(k-fold cross validation): 데이터셋을 k개로 나누어 하나씩 테스트셋으로 사용하고 나머지 모두를 학습셋으로 사용
    - ex> 100개 샘플의 전체 데이터셋을 5개로 나누는 5겹 교차 검증
        1. 100개의 샘플을 20개씩 5개 셋으로 나눔
        2. 1, 2, 3, 4번 셋을 학습하고 5번 셋으로 테스트한 결과 1 계산
        3. 1, 2, 3, 5번 셋을 학습하고 4번 셋으로 테스트한 결과 2 계산
        4. 1, 2, 4, 5번 셋을 학습하고 3번 셋으로 테스트한 결과 3 계산
        5. 1, 3, 4, 5번 셋을 학습하고 2번 셋으로 테스트한 결과 4 계산
        6. 2, 3, 4, 5번 셋을 학습하고 1번 셋으로 테스트한 결과 5 계산
        7. 결과 1~5의 평균이 최종 결과

```python
from sklearn.model_selection import KFold

# 몇 겹으로 나눌 것인지 정합니다.
k = 5

# KFold 함수를 불러옵니다. 분할하기 전에 샘플이 치우치지 않도록 섞어 줍니다.
kfold = KFold(n_splits=k, shuffle=True)

# 정확도가 채워질 빈 리스트를 준비합니다.
acc_score = []
```

```python
def model_fn():
    model = Sequential()  # 딥러닝 모델의 구조를 시작합니다.
    model.add(Dense(24, input_dim=60, activation="relu"))
    model.add(Dense(10, activation="relu"))
    model.add(Dense(1, activation="sigmoid"))
    return model
```

```python
# k겹 교차 검증을 이용해 k번의 학습을 실행합니다.
# for 문에 의해 k번 반복합니다.
# split()에 의해 k개의 학습셋, 테스트셋으로 분리됩니다.
for train_index, test_index in kfold.split(X):
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model = model_fn()
    model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
    history = model.fit(X_train, y_train, epochs=200, batch_size=10, verbose=0)

    accuracy = model.evaluate(X_test, y_test)[1]  # 정확도를 구합니다.
    acc_score.append(accuracy)  # 정확도 리스트에 저장합니다.
```

```python
# k번 실시된 정확도의 평균을 구합니다.
avg_acc_score = sum(acc_score) / k

# 결과를 출력합니다.
print("정확도:", acc_score)
print("정확도 평균:", avg_acc_score)
```
